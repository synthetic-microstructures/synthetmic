In this example, we will show how to use `synthetmic` package to generate
a 3D synthetic microstructure using Laguerre diagrams.

Since this is an introduction, we will create a 3D Laguerre diagram
with all cells or grains having the same volumes.

We first install `synthetmic`.

```python
!pip install synthetmic
```

Next, we import the generator class and functions.

```python
from synthetmic import LaguerreDiagramGenerator
from synthetmic.data.toy import create_data_with_constant_volumes
from synthetmic.plot import plot_cells_as_pyvista_fig
```

We now create data for the diagram. Below, we create a unit cube with 100 grains.
All grains will have the volume $1/100$ cubic units.

Note that you can always create your own custom data and not required to
use the provided functions in `synthetmic`.

```python
data = create_data_with_constant_volumes(
    space_dim=3,
    n_grains=100,
    is_periodic=False,
    random_state=42
)
data
```

We then create an instance of `LaguerreDiagramGenerator` object, and call `fit` method on the created data.

```python
diagram = LaguerreDiagramGenerator(
    tol=1.0,
    n_iter=5,
    damp_param=1.0,
    verbose=True,
)

diagram.fit(
    seeds=data.seeds,
    volumes=data.volumes,
    domain=data.domain,
    periodic=data.periodic,
    init_weights=data.init_weights
)
```

Once the diagram is created, we can access many properties.
For instance, the fitted volumes, vertices, centroids, mesh, etc.

```python
diagram.get_fitted_volumes()
```

```python
diagram.get_vertices()
```

```python
diagram.get_centroids()
```

```python
diagram.get_mesh()
```

With the help of plotting function, we can plot the generated diagram.

```python
pl = plot_cells_as_pyvista_fig(generator=diagram, colorby=diagram.get_fitted_volumes(), notebook=True)
pl.show()
```

If you want periodicity in any coordinate (say all coordinates),
we just need to modify the `fit` method call to use
`periodic=[True, True, True]` as follows:

```python
diagram.fit(
    seeds=data.seeds,
    volumes=data.volumes,
    domain=data.domain,
    init_weights=data.init_weights,
    periodic=[True, True, True]
)
```

We can then plot the diagram accordingly.

```python
pl = plot_cells_as_pyvista_fig(generator=diagram, colorby=diagram.get_fitted_volumes(), notebook=True)
pl.show()
```
